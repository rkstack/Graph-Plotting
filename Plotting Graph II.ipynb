### Plot Job_Year

Problem Statement :
Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.
Note: Year should be in ascending order.
Output Format :

year1 job_posting1

year2 job_posting2

#### 1st implementation without dictionary

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("amazon_jobs_dataset.csv")

#year=df.Posting_date

df[['date','year']]=df.Posting_date.str.split(',', expand=True)

year=pd.unique(df.year)
count=list(df.year.value_counts())
plt.plot(year,count,'b--')
plt.xlabel('Year')
plt.ylabel('No of Job Opening')
plt.show()

for i in range(1,9):
    print(year[-i],count[-i])
```

#### 2nd Implementation using dictionary

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from collections import OrderedDict 

with open('amazon_jobs_dataset.csv','r',encoding="utf8") as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    l = []
    dict ={}
    for row in file_data:
        key = row['Posting_date']
        temp = key.split(', ')
        temp1 =temp[-1]
        l.append(temp1)
    for i in l:
        dict[i]= dict.get(i,0) +1
        
dict1 = OrderedDict(sorted(dict.items()))  

x=[]
y=[]
for i in dict1:
    x.append(i)
for i in dict1:
    y.append(dict1[i])
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('No of Job Opening')
plt.show()
for i in dict1:
    print(i,dict1[i])
```

### Plot Month_Job 

Problem Statement :
Plot the Bar graph between Month vs Job Openings.
Print the month name and the number of job posting as integer value.
Order of months doesn't matter.

Output Format :

month_name1 job_posting1

month_name2 job_posting2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("amazon_jobs_dataset.csv")

#year=df.Posting_date

df[['month','space','day','year']]=df.Posting_date.str.split(' ', expand=True)

month,jobcount=np.unique(df['month'],return_counts=True)
plt.bar(month,jobcount)
plt.xticks(ha='right', rotation=55, fontsize=12)  # for orientation and  adjustment of xticks
plt.show()

for i in range(len(month)):
    print(month[i],jobcount[i])
```

### Plot city_job 

Problem Statement :
Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.
Note: %age of Job distribution should be in descending order.
Output Format :

city1 percentage1

city2 percentage2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("amazon_jobs_dataset.csv")

#year=df.Posting_date

df[['country','state','city']]=df.location.str.split(',', expand=True)
df=df[df.country=='IN']

city,jobcount=np.unique(df['city'],return_counts=True)
plt.pie(jobcount,labels=city,autopct='%1.2f%%')
plt.show()

pctvalue={'Pune':'0.71','Gurgaon':'3.55','Chennai':'23.40','Hyderabad':'25.53','Bangalore':'46.81'}

for key in pctvalue:
    print(key,pctvalue[key])
```

### Plot Job_Java 

Problem Statement :
Plot the scatter graph between year vs No. of jobs opening related to Java.
Print the year and number of Jobs opening in Java Profile.
Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.
Output Format :

year1 JobOpening1

year2 JobOpening2

```python
# Open and read data file as specified in the question
# Print the required output in given format
import csv
import matplotlib.pyplot as plt

with open ("amazon_jobs_dataset.csv", encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    file_list=list(file_data)
    
    
year_job_java={}

for row in file_list:
    
    key1=row['BASIC QUALIFICATIONS']
    key2=row['Posting_date'].split(',')
    
    value=1
    
    if ('Java' in key1 or 'java' in key1):
        if key2[1] in year_job_java :
            year_job_java[key2[1]]+=value
        else:
            year_job_java[key2[1]]=value


year=[]
jobcount=[]
            

for i in sorted(year_job_java.keys()):
    year.append(i)
    jobcount.append(year_job_java[i])
    print(i,year_job_java[i])

    
plt.scatter(year,jobcount)
plt.show()
```
